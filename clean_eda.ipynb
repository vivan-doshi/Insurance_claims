```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plot
plt.style.use('ggplot')
sns.set(font_scale=1.1)

# Load the data
df = data.copy()

# Define age and power bins
age_bins = [18, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80]
age_labels = [f"{age_bins[i]}-{age_bins[i+1]-1}" for i in range(len(age_bins)-1)]

power_bins = [0, 50, 100, 150, 200, 250, 300, 350, 400]
power_labels = [f"{power_bins[i]}-{power_bins[i+1]-1}" for i in range(len(power_bins)-1)]

# Create age and power categories
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
df['Power_Group'] = pd.cut(df['Vehicle_Power_HP'], bins=power_bins, labels=power_labels, right=False)

# Group by age and power to calculate claim frequency
claim_freq = df.groupby(['Age_Group', 'Power_Group'])['Claim_Status'].agg(['sum', 'count']).reset_index()
claim_freq['frequency'] = claim_freq['sum'] / claim_freq['count']

# Reshape for heatmap
heatmap_data = claim_freq.pivot(index='Age_Group', columns='Power_Group', values='frequency')

# Calculate average claim frequency for reference
avg_frequency = df['Claim_Status'].mean()
print(f"Average claim frequency: {avg_frequency:.2%}")

# Create the heatmap
plt.figure(figsize=(14, 10))
ax = sns.heatmap(heatmap_data, 
                 annot=True, 
                 fmt='.1%', 
                 cmap='crest',  # Red-Yellow-Green reversed (Red = high risk)
                 linewidths=0.5,
                 vmin=0, 
                 vmax=0.3,  # Cap at 30% for better color distribution
                 cbar_kws={'label': 'Claim Frequency'})

# Customize the heatmap
plt.title('Claim Frequency by Driver Age and Vehicle Power', fontsize=16, pad=20)
plt.figtext(0.5, 0.01, 'Risk concentration in young drivers with high-powered vehicles', 
            ha='center', fontsize=12, style='italic')
plt.figtext(0.5, 0.95, f'Average claim frequency across all policies: {avg_frequency:.1%}', 
            ha='center', fontsize=10)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# Add labels
ax.set_xlabel('Vehicle Power (HP)', fontsize=12, labelpad=10)
ax.set_ylabel('Driver Age (years)', fontsize=12, labelpad=10)

# Add custom legend for risk levels
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='#d73027', label='High Risk (≥25%)'),
    Patch(facecolor='#fc8d59', label='Med-High Risk (15-25%)'),
    Patch(facecolor='#fee08b', label='Medium Risk (10-15%)'),
    Patch(facecolor='#d9ef8b', label='Low-Med Risk (5-10%)'),
    Patch(facecolor='#91cf60', label='Low Risk (<5%)')
]
plt.legend(handles=legend_elements, loc='upper right', bbox_to_anchor=(1.15, 0.2),
           title='Risk Levels', frameon=True)

# Save the figure
plt.tight_layout()
plt.savefig('claim_frequency_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

# Print out some statistics about the highest risk combinations
high_risk = claim_freq.sort_values('frequency', ascending=False).head(5)
print("\nTop 5 highest risk combinations:")
print(high_risk[['Age_Group', 'Power_Group', 'frequency', 'count']])
```
